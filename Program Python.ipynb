```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style to white to remove background color
sns.set(style="white")

# Daftar fitur yang dipilih
fitur_terpilih = ['Age', 'Weight', 'Height', 'Height/Age ZS', 'Weight/Age ZS', 'Weight/Height ZS']
df_terpilih = df[fitur_terpilih]

# Filter data untuk menghapus baris dengan 'Usia Saat Ukur' sama dengan '0 Tahun - 0 Bulan - 0 Hari'
df_terpilih = df_terpilih[df_terpilih['Age'] != '0 Tahun - 0 Bulan - 0 Hari']

# Reset index
df_terpilih.reset_index(drop=True, inplace=True)

# Fungsi untuk mengkonversi 'Usia Saat Ukur' menjadi bulan
def convert_age_to_months(age_str):
    try:
        parts = age_str.split('-')
        year = int(parts[0].strip().split()[0])
        month = int(parts[1].strip().split()[0])
        day = int(parts[2].strip().split()[0])
        return year * 12 + month + day / 30
    except ValueError:
        return None

# Konversi 'Usia Saat Ukur' ke bulan
df_terpilih['Age'] = df_terpilih['Age'].apply(convert_age_to_months)

# Ubah seluruh DataFrame menjadi numerik, hilangkan kolom dengan semua NaN
df_numerik = df_terpilih.apply(pd.to_numeric, errors='coerce')
df_numerik = df_numerik.dropna(axis=1, how='all')

# Fungsi untuk melakukan Windsorization (mengganti outlier dengan batas persentil)
def windsorize(df, column, lower_percentile=0.05, upper_percentile=0.95):
    lower_bound = df[column].quantile(lower_percentile)
    upper_bound = df[column].quantile(upper_percentile)
    df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
    return df

# Terapkan Windsorization untuk setiap fitur yang dipilih
for column in fitur_terpilih:
    if column in df_numerik.columns:
        df_numerik = windsorize(df_numerik, column)

# Membuat box plots
plt.figure(figsize=(9, 8))

for i, column in enumerate(fitur_terpilih, 1):
    plt.subplot(2, 3, i)
    if column in df_numerik.columns:
        sns.boxplot(y=df_numerik[column], color='white', linewidth=1.2, boxprops=dict(facecolor='white', edgecolor='black'))  # Set box plot color to white
        plt.title(f'{column}')
        plt.ylabel('')  # Remove y-axis label
    else:
        plt.title(f'No data for {column}')
    plt.xticks([])  # Hide x-axis ticks

    # Remove background and grid lines
    plt.gca().patch.set_facecolor('white')
    plt.grid(False)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['left'].set_visible(False)
    plt.gca().spines['bottom'].set_visible(False)

    # Add a box around each plot
    plt.gca().add_patch(plt.Rectangle((0, 0), 1, 1, transform=plt.gca().transAxes, color='black', fill=False, linewidth=1.2))

plt.tight_layout()
plt.show()
```

IMPLEMENTASI KNN

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, f1_score
from sklearn.preprocessing import LabelEncoder

# Mengubah format kolom 'Age' menjadi jumlah total bulan
def convert_age_to_months(age_str):
    parts = age_str.split()
    years = int(parts[0]) * 12
    months = int(parts[3])
    days = int(parts[6]) / 30.0
    total_months = years + months + days
    return total_months

# Asumsi df sudah ada sebelumnya
df['Usia Bulan'] = df['Age'].apply(convert_age_to_months)

# Menambahkan 'Usia Bulan' ke dalam fitur
fitur = ['Usia Bulan','Weight', 'Height', 'Height/Age ZS', 'Weight/Age ZS', 'Weight/Height ZS']
X = df[fitur]
y = df['Height/Age']  # Kolom target yang ingin diprediksi

# Mengonversi target y ke nilai numerik
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
class_names = label_encoder.classes_.astype(str)

# Membagi data menjadi set latih dan set uji
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Membuat objek KNN dengan k=1
knn_model = KNeighborsClassifier(n_neighbors=1)

# Melatih model KNN dengan data latih
knn_model.fit(X_train, y_train)

# Prediksi menggunakan data uji
y_pred = knn_model.predict(X_test)

# Menghitung akurasi pada data uji
acc_knn = accuracy_score(y_test, y_pred)
print(f"Accuracy Score: {acc_knn*100:.2f}%\n")
precision_knn = precision_score(y_test, y_pred, average='weighted') * 100
recall_knn = recall_score(y_test, y_pred, average='weighted') * 100
f1_score_knn = f1_score(y_test, y_pred, average='weighted') * 100

# Menampilkan nilai presisi, recall, dan F1 score
print(f"Presisi KNN: {precision_knn:.2f}%")
print(f"Recall KNN: {recall_knn:.2f}%")
print(f"F1 Score KNN: {f1_score_knn:.2f}%")

# Classification Report pada data uji
print("Classification Report for KNN Algorithm: ")
print(classification_report(y_test, y_pred, target_names=class_names))

# Confusion Matrix untuk KNN pada data uji
conf_matrix_knn = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_knn, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix - K-Nearest Neighbors (KNN)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, KFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, f1_score
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE
from sklearn.impute import SimpleImputer

# Mengubah format kolom 'Age' menjadi jumlah total bulan
def convert_age_to_months(age_str):
    parts = age_str.split()
    years = int(parts[0]) * 12
    months = int(parts[3])
    days = int(parts[6]) / 30.0
    total_months = years + months + days
    return total_months

# Asumsi df sudah ada sebelumnya
df['Usia Bulan'] = df['Age'].apply(convert_age_to_months)

# Menambahkan 'Usia Bulan' ke dalam fitur
fitur = ['Usia Bulan', 'Weight', 'Height', 'Height/Age ZS', 'Weight/Age ZS', 'Weight/Height ZS']
X = df[fitur]
y = df['Height/Age']  # Kolom target yang ingin diprediksi


# Mengisi nilai yang hilang dengan mean kolom terkait
imputer = SimpleImputer(strategy='mean')
X = imputer.fit_transform(X)

# Label Encoder untuk target
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Membagi data menjadi set latih dan set uji
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.4, random_state=42)

# Inisialisasi model K-Fold Cross Validation dengan k=5
k_fold = KFold(n_splits=5, shuffle=True, random_state=42)

# Menyimpan skor dari setiap fold
cv_scores = []
for train_index, val_index in k_fold.split(X_train):
    X_train_fold, X_val_fold = X_train[train_index], X_train[val_index]
    y_train_fold, y_val_fold = y_train[train_index], y_train[val_index]

    # Terapkan SMOTE pada data latih
    smote = SMOTE(random_state=42)
    X_train_resampled, y_train_resampled = smote.fit_resample(X_train_fold, y_train_fold)

    # Membuat objek KNN dengan k=1
    knn_model = KNeighborsClassifier(n_neighbors=1)

    # Melatih model KNN dengan data latih yang telah di-oversample
    knn_model.fit(X_train_resampled, y_train_resampled)

    # Evaluasi model pada data validasi
    y_pred = knn_model.predict(X_val_fold)
    acc_fold = accuracy_score(y_val_fold, y_pred)
    cv_scores.append(acc_fold)

# Menampilkan hasil akurasi dari K-Fold Cross Validation
print("Cross Validation Scores:", cv_scores)
print("Mean Accuracy:", sum(cv_scores) / len(cv_scores))

# Melatih model KNN dengan seluruh data latih yang telah di-oversample
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Imputasi untuk mengisi nilai NaN pada data uji
X_test = imputer.transform(X_test)

# Membuat objek KNN dengan k=1
knn_model = KNeighborsClassifier(n_neighbors=1)

# Melatih model KNN dengan data latih yang telah di-oversample
knn_model.fit(X_train_resampled, y_train_resampled)

# Prediksi menggunakan data uji
y_pred = knn_model.predict(X_test)

# Menghitung akurasi pada data uji
acc_knn = accuracy_score(y_test, y_pred)
print(f"\nAccuracy Score: {acc_knn*100:.2f}%")

# Menampilkan nilai presisi, recall, dan F1 score pada data uji
precision_knn = precision_score(y_test, y_pred, average='weighted') * 100
recall_knn = recall_score(y_test, y_pred, average='weighted') * 100
f1_score_knn = f1_score(y_test, y_pred, average='weighted') * 100
print(f"Presisi KNN: {precision_knn:.2f}%")
print(f"Recall KNN: {recall_knn:.2f}%")
print(f"F1 Score KNN: {f1_score_knn:.2f}%")

# Classification Report pada data uji
print("\nClassification Report for KNN Algorithm: ")
print(classification_report(y_test, y_pred))

# Confusion Matrix for KNN pada data uji
conf_matrix_knn = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_knn, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.title("Confusion Matrix - K-Nearest Neighbors (KNN)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
